## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
!pip install pandas numpy
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000

data = pd.DataFrame({
    "emergency_admissions": np.random.poisson(20, n),
    "scheduled_admissions": np.random.poisson(15, n),
    "discharges": np.random.poisson(18, n),
    "flu_cases": np.random.poisson(8, n),
    "occupancy_rate": np.random.normal(75, 10, n),
    "available_staff": np.random.randint(25, 50, n)
})

data["bed_demand"] = (
    data["emergency_admissions"]
    + data["scheduled_admissions"]
    - data["discharges"]
    + data["flu_cases"]
)

print("Number of observations:", len(data))
print("Average daily bed demand:", round(data["bed_demand"].mean(), 2))
print("Maximum observed bed demand:", data["bed_demand"].max())
print("Average occupancy rate:", round(data["occupancy_rate"].mean(), 2))
```
